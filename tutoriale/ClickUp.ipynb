# ClickUp - tutorial

Celem mojego zadania jest przedstawienie tajników narzędzia ClickUp, które mogą okazać się pomocne podczas pracy przy naszym projekcie.
W poradniku skupię się głównie na elementach widocznych na pasku bocznym znajdującym się po lewej stronie ekranu i krótkim opisie ich najistotniejszych składowych.


Pierwszą zakładką jest zakładka Home.

Dostępnymi tu funkcjonalnościami są:

* LineUp – tworzący zhierarchizowaną listę zadań przypisanych zarówno do nas, jak i innych członków zespołu (musimy dodać zadanie do listy, nowo powstałe zadania nie pojawiają się tu automatycznie);
* Trending – lista ostatnio wyświetlanych zadań;
* My Work – zawiera listę wszystkich przewidzianych dla nas zadań (To do), które są zgrupowane według ich ustalonego (bądź nie) terminu wykonania (Today, Overdue, Next, Unscheduled). Ponadto możemy zobaczyć zadania, które udało nam się zakończyć (Done) oraz osoby, którym ustawiliśmy „przypominajkę” (Delegated) (nie musi ona dotyczyć konkretnego zadania).

Po prawej stronie widoczne są kalendarz oraz terminarz, których rola jest intuicyjna. 
Mentions zawiera wszystkie wzmianki, które ktokolwiek poczynił o nas w komentarzach.

Kolejną zakładką jest zakładka Notifications, w której pojawiają się powiadomienia (co raczej nie jest ogromnym zaskoczeniem), dotyczące aktywności związanych z przypisanymi nam zadaniami. Tak to mniej więcej wygląda:

Zakładka Pulse dostarcza nam informacji o aktywności członków zespołu, zaś w zakładce Goals możemy tworzyć cele, które zamierzamy zrealizować (nie jest to raczej na tyle istotne, by dokładniej tłumaczyć jak to wykonać, dla zainteresowanych: https://www.youtube.com/watch?v=QZmwLZ-kUU0&t=232s&ab_channel=MatchaMedia (sygnatura czasowa 13:33-15:25)).

Przejdę teraz do omówienia zakładki Spaces, która prawdopodobnie będzie najczęściej eksplorowana i to jej opisowi poświęcę najwięcej uwagi.

Nasza „przestrzeń”, która została już utworzona kryje się pod etykietą „Big Mess”.  Aby stworzyć analogiczną przestrzeń projektową, klikamy „+ New space”. Ukaże nam się okno, w którym wybieramy nazwę dla naszego projektu, rodzaje statusów, jakie może osiągnąć zadanie oraz elementy wizualne awataru. Po przejściu wszystkich kroków klikamy „Create Space”, potwierdzając nasze wybory i tym samym tworząc nową przestrzeń.

Na przestrzeń składają się foldery, które grupują zadania całego zespołu. Nowy folder możemy dodać przez naciśnięcie „+” przy nazwie projektu. Wewnątrz folderów znajdują się listy, których kolejność uszeregowuje prace nad projektem (analogicznie, możemy dodać listę do folderu przez naciśnięcie „+” przy jego nazwie). Wchodząc w dowolną listę widzimy na jakim etapie są postępy w pracy. 

Ostatnim elementem przestrzeni jest Backlog, który stanowi zbiór zadań potrzebnych do zamknięcia projektu.

Skupimy się teraz na funkcjonalnościach jakie wchodzą w skład naszej przestrzeni projektowej. 

Na górnym pasku mamy automatycznie ustawioną kartę „List”. Przechodząc na kartę „Board” uzyskujemy te same informacje co z karty poprzedniej, jednak przedstawione z wykorzystaniem tablicy Kanban.

Możliwości jakie dają nam funkcje dostępne w widoku konkretnej listy mają charakter czysto praktyczny. Celem powstania listy jest przecież utworzenie zadań, przypisanie ich członkom zespołu, a następnie sukcesywne ich wykonywanie aż do stanu „Done”. I na każdym z tych elementów się skupimy.

##### Tworzenie zadań

Raczej nie powinno stanowić problemu, klikamy „+ New task” znajdujący się na samym dole zbioru zadań z dowolnego stanu, a następnie nadajemy mu nazwę (naturalnym jest dodawanie zadań do stanu „Open”, choć jak widzimy możemy to zrobić wszędzie)

##### Czynności związane z utworzeniem zadania (i jak je zrobić):

* przypisanie do niego wykonawcy (kolor czerwony);
* możliwość ustalenia (bądź nie) jego istotności (kolor żółty);
* możliwość dodania opisu (kolor zielony);
* możliwość zadeklarowania podzadań (kolor niebieski);
* możliwość dodania powiązanych zadań, dokumentów itp. (kolor różowy);
* możliwość utworzenia listy kontrolnej (kolor pomarańczowy);

* możliwość dodania szacunkowego czasu potrzebnego do ukończenia zadania (kolor fioletowy);

* możliwość dodania terminu wykonania zadania (kolor brązowy);

* możliwość dodania tzw. punktów sprintu, czyli mierzonej wartością liczbową ilości wysiłku potrzebnego do wykonania zadania (kolor szary).

##### Czynności związane z wykonywaniem zadania (i jak je zrobić):

* zmiana stanu zadania (kolor czerwony);
* tworzenie komentarzy/zapytań związanych z zadaniem (kolor żółty);
* dodawanie załączników do zadań (kolor zielony);
* możliwość dodania mierzenia czasu wykonywania zadania (po uprzednim połączeniu ClickUp z Clockify) (kolor niebieski).
